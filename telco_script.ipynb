## Nonlinearity

I introduce nonlinearities for linear models - logistic regression and Cox proportional hazard model.

```python
df_dummy['MonthlyCharges_log'] = np.log(df_dummy['MonthlyCharges'])
df_dummy['MonthlyCharges_pow_2'] = np.power(df_dummy['MonthlyCharges'], 2)
df_dummy['tenure_log'] = np.log(df_dummy['tenure'])
df_dummy['tenure_pow_2'] = np.power(df_dummy['tenure'], 2)
```

# Survival plots

In this section I will present survival plots for selected variables, which present survival ratio (percent of non churned clients) for particular features on the timeline.

```python
kmf = lifelines.KaplanMeierFitter() 

no_internet_service = df_dummy['InternetService_No']   
no_internet = no_internet_service == 1

T = df_dummy['tenure']
C = df_dummy['Churn']
kmf.fit(T[~no_internet], C[~no_internet], label='Internet services')
a1 = kmf.plot()

kmf.fit(T[no_internet], C[no_internet], label='No internet')
kmf.plot(ax=a1)
```

With the survival plot above we can see that those customers without internet service are much more loyal and are reluctant to leave company. This phenomenon should be checked by the company, for it may seem that the sector of internet services is very competitive and may not be profitable.

```python
seniors = df_dummy['SeniorCitizen']   
senior = seniors == 1

T = df_dummy['tenure']
C = df_dummy['Churn']
kmf.fit(T[~senior], C[~senior], label='No seniors')
a1 = kmf.plot()

kmf.fit(T[senior], C[senior], label='Seniors')
kmf.plot(ax=a1)
```

The above plot indicates that senior citizens are less loyal and leave company earilier than no seniors.

```python
# Survival plot for people with internet and tech support
no_internet = df_dummy['InternetService_No']  
tech = df_dummy['TechSupport_Yes']
tech_support = (no_internet == 0) & (tech == 1)
no_tech_support = (no_internet == 0) & (tech == 0)

T = df_dummy['tenure']
C = df_dummy['Churn']
kmf.fit(T[no_tech_support], C[no_tech_support], label='No tech support')
a1 = kmf.plot()

kmf.fit(T[tech_support], C[tech_support], label='Tech support')
kmf.plot(ax=a1)
```

On the next plot one can see that customers with tech support are more loyal. It is not known from the dataset why one customers have tech support and others do not have, but I guess that tech support is a premium service and it helps to retain clients.

```python
# Survival plot for people with internet and tech support
one_year_contract = df_dummy['Contract_One year']  
two_year_contract = df_dummy['Contract_Two year']  
monthly_contract = (one_year_contract == 0) & (two_year_contract == 0)
one_year_contract = one_year_contract == 1
two_year_contract = two_year_contract == 1

T = df_dummy['tenure']
C = df_dummy['Churn']
kmf.fit(T[monthly_contract], C[monthly_contract], label='Monthly contract')
a1 = kmf.plot()
kmf.fit(T[one_year_contract], C[one_year_contract], label='One year contract')
a2 = kmf.plot(ax=a1)

kmf.fit(T[two_year_contract], C[two_year_contract], label='Two year contract')
kmf.plot(ax=a2)
```

With the contract time survival plot we can see that long-term contracts retains clients for longer, even for much longer than duration of the contract. While the short-term contract makes customers churn.

# Models

## Data partition and oversampling (SMOTE)

```python
X = df_dummy.drop(columns=['Churn'])
y = df_dummy['Churn'].values
os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
columns = X_train.columns

os_data_X, os_data_y = os.fit_sample(X_train, y_train)
os_data_X = pd.DataFrame(data=os_data_X, columns=columns)
os_data_y = pd.DataFrame(data=os_data_y, columns=['Churn'])
# we can Check the numbers of our data
print("length of oversampled data: ", len(os_data_X))
print("Number of no churn in oversampled data: ", len(os_data_y[os_data_y['Churn'] == 0]))
print("Number of churn: ", len(os_data_y[os_data_y['Churn'] == 1]))
print("Proportion of no churn data in oversampled data is: ", len(os_data_y[os_data_y['Churn']==0]) / len(os_data_X))
print("Proportion of churn data in oversampled data is: ", len(os_data_y[os_data_y['Churn']==1]) / len(os_data_X))

print(os_data_X.shape)
print(X_test.shape)
os_data_y.shape
```

## Logistic regression

```python
log_train = X_train.copy()
log_train['intercept'] = 1.0
logistic_regression = sm.Logit(y_train, log_train)
result = logistic_regression.fit()
print(result.summary())
```

Obtained results indicate that higher risk of churn reveals new clients than old ones. One may see that both statistically significant time variables, tenure_log and tenur_pow_2, have negative coefficients. The same conclusion may be arrived at after analysis of the tenure denisty plot, presented in visualisation section.

## Cox proportional hazard model

```python
cox = lifelines.CoxPHFitter()
y_df = pd.DataFrame(data=y_train, columns=['Churn'])
cox_df = pd.concat([X_train.reset_index(), y_df], axis=1).copy()
cox_df.drop(columns=['index', 'tenure_log', 'tenure_pow_2'], inplace=True)

cox.fit(cox_df, duration_col='tenure', event_col='Churn')
cox.print_summary()
```

The Cox model is the only one where monthly charges are not important in explaining churn probability.

# Predictions

```python
model_logistic = linear_model.LogisticRegression()
model_logistic.fit(os_data_X, os_data_y.values.ravel())
logistic_pred = model_logistic.predict(X_test)

model_logistic_unbalanced = linear_model.LogisticRegression()
model_logistic_unbalanced.fit(X_train, y_train.ravel())
logistic_pred_unbalanced = model_logistic_unbalanced.predict(X_test)

# ROC curve
logit_roc_auc = roc_auc_score(y_test, model_logistic.predict_proba(X_test)[:, 1])
logit_unbalanced_roc_auc = roc_auc_score(y_test, model_logistic_unbalanced.predict_proba(X_test)[:, 1])
fpr, tpr, thresholds = roc_curve(y_test, model_logistic.predict_proba(X_test)[:, 1])
fpr_unb, tpr_unb, thresholds_unb = roc_curve(y_test, model_logistic_unbalanced.predict_proba(X_test)[:, 1])

plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot(fpr_unb, tpr_unb, label='Logistic Regression unbalanced (area = %0.2f)' % logit_unbalanced_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

Oversampling turned out to have no effect on the performance of logistic regression.

```python
cox_pred = cox.predict_survival_function(X_test.iloc[0:1,:], times=range(1, 73, 1))
cox_pred

def predict_cox(cox_model, test_data, time_var: str = 'tenure'):
    cox_pred = []
    i = 0
    for index, row in test_data.iterrows():
        tenure = row[time_var]
        prob = 1.0 - cox_model.predict_survival_function(test_data.iloc[i:(i+1),:], 
                                                                  times=[tenure])
        cox_pred.append(prob)
        i += 1
    return pd.DataFrame(data=cox_pred, columns=['probability'])

cox_pred = predict_cox(cox, X_test, 'tenure')
```

```python
cox_roc_auc = roc_auc_score(y_test, cox_pred)
fpr_cox, tpr_cox, thresholds_cox = roc_curve(y_test, cox_pred)
```

## XGBoost

```python
# Elimination of unnecessary variables - XGBoost as any tree based method does not need any kind of 
# nonlinear transformation
xgb_train = X_train.copy()
xgb_train.drop(columns=['tenure_log', 'tenure_pow_2', 
                     'MonthlyCharges_log', 'MonthlyCharges_pow_2'], inplace=True)
xgb_test = X_test.copy()
xgb_test.drop(columns=['tenure_log', 'tenure_pow_2', 
                     'MonthlyCharges_log', 'MonthlyCharges_pow_2'], inplace=True)

# Selecting validation set
dXtrain, dXvalid, dYtrain, dYvalid = train_test_split(xgb_train, y_train, test_size=0.2, random_state=2)
# Transforming pandas df to xgboost format
dtrain = xgb.DMatrix(dXtrain, label=dYtrain)
dvalid = xgb.DMatrix(dXvalid, label=dYvalid)
dtest = xgb.DMatrix(xgb_test, label=y_test)
```

```python
# Parameters' tuning
param_test = {
    'max_depth': range(3, 8, 2),
    'min_child_weight': range(1, 6, 2),
    'learning_rate': [0.01, 0.05, 0.1],
    'gamma': [0, 5, 10]
}
gsearch = GridSearchCV(estimator = XGBClassifier(learning_rate=0.1, n_estimators=100, max_depth=5,
                                                 min_child_weight=1, gamma=0, subsample=0.8, 
                                                 colsample_bytree=0.8, objective='binary:logistic', 
                                                 n_jobs=4, scale_pos_weight=1, random_state=1), 
                       param_grid = param_test, scoring='roc_auc', n_jobs=4, iid=False, cv=3)
gsearch.fit(X_train, y_train)
gsearch.best_params_, gsearch.best_score_
```

```python
# Training with optimised parameters
param = {'max_depth': 3, 'eta': 0.05, 'objective': 'binary:logistic', 'nthread': 3, 'eval_metric': 'auc',
        'gamma': 0, 'seed': 1}
evallist = [(dtrain, 'train'), (dvalid, 'eval')]

num_round = 100
model_xgb = xgb.train(param, dtrain, num_round, evallist, early_stopping_rounds=10)
```

```python
# Prediction 
xgb_pred = model_xgb.predict(dtest, ntree_limit=bst.best_ntree_limit)

xgb_roc_auc = roc_auc_score(y_test, xgb_pred)
fpr_xgb, tpr_xgb, thresholds_xgb = roc_curve(y_test, xgb_pred)
```

```python
xgb.plot_importance(model_xgb)
```

Plot of feature importance says that the most important in predicting churn are length of tenure and value of monthly charges. Less important, but still much more than the rest of variables, is the contract for two year, as it was shown on survival plot.

### Clasification

```python
churn_ratio = np.mean(y_train)
print('Churn ratio: ', str(round(churn_ratio, 3)))
# XGB
print('XGBoost: ')
xgb_churn_pred = (xgb_pred > churn_ratio)
xgb_churn_pred.astype('int')

print(confusion_matrix(y_test, xgb_churn_pred))
print(classification_report(y_test, xgb_churn_pred))

# Logistic regression
# Unbalanced
print('Unbalanced logistic regression: ')
logistic_churn_pred = (model_logistic_unbalanced.predict_proba(X_test)[:, 1] > churn_ratio)
logistic_churn_pred.astype('int')

print(confusion_matrix(y_test, logistic_churn_pred))
print(classification_report(y_test, logistic_churn_pred))

# Balanced
print('Balanced logistic regression: ')
logistic_churn_pred = (model_logistic.predict_proba(X_test)[:, 1] > 0.5)
logistic_churn_pred.astype('int')

print(confusion_matrix(y_test, logistic_churn_pred))
print(classification_report(y_test, logistic_churn_pred))

# Cox
print('Cox model: ')   
cox_churn_pred = (cox_pred > churn_ratio)
cox_churn_pred.astype('int')

print(confusion_matrix(y_test, cox_churn_pred))
print(classification_report(y_test, cox_churn_pred))
```

# Summary
Logistic regression and XGBoost provided similar results. However, if I were to choose only one model, I would opt for XGBoost, because it seems to be slightly better in recall of churn.


Nonetheless, the final model, as well as the probability cut-off for classification of an observation as churn should be optimised with the use of relative churn cost (value of the client vs cost of promotion or any other way of keeping customer with the company or service).


Far worse than XGBoost and logistic regression turned out to be Cox proportional hazard model. 

Statistical analysis showed that long-term contracts are associated with greater fidelity, however, it is not certain whether the long-term contracts make clients reluctant to churn or clients that are reluctant to change their service provider opt for longer contracts. Nonetheless, the fact is that old clients are less probable to churn than the new ones.
